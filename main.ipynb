## Stochastic Block Models

In these kinds of models, nodes fall under different groups and the probability of having within-ties and between-ties in the groups would be independent conditionally on the group to which nodes belong. 




## Latent Models

These sets of models fall under Random graphs and try to assign the probability of having a tie between two nodes by measuring the distance between that two nodes in some latent space. In these models the probability of two nodes sharing a tie depends on the covariates of the nodes and the distance between them in terms of some statistic. An example of these models can ignore the covariates and only depend on the distance between two nodes, two nodes share a tie if and only if their distance is less than 1. We call a network, d-k representable if we could find points in a k-dimensional space for every node in the network that they can satisfy the previously mentioned condition having distance less than one.




## Transitivity

This is a structure which is defined in social networks in a way that if i is connected to j and j is connected to k, then transitivity measures the degree to which the edge from i to k is probable.



## Reciprocity

Reciprocity is defined as the tendency of node A to have edge to node B if B, already has a edge to A. 



## Homophily

Homophily refers to the tendency for people to have (non-negative) ties with people who are similar to themselves in socially significant ways. The term itself (due to Lazarsfeld) specifically refers to an internal preference.

### Sequence to Sequence Models

A sequence-to-sequence model is a model that takes a sequence of items (words, letters, features of an imagesetc) and outputs another sequence of items. Under the hood, the model is composed of an encoder and a decoder.

The encoder processes each item in the input sequence, it compiles the information it captures into a vector (called the context). After processing the entire input sequence, the encoder sends the context over to the decoder, which begins producing the output sequence item by item.

Attention allows the model to focus on the relevant parts of the input sequence as needed. 

By getting all the hidden states produced at the encoding phase, in decoding phase, at each time step, a score is assigned to each of the hidden states(softmax) and the weighted sum of the hidden states would give us the context vector at that time step in decoding phase.

In each time step at the decoding phase, the RNN would get the hidden state from the last time step and the input word embedding and the context vector would be resulted from the hidden states came from the encoding phase and the hidden state produced in the current time step. A FeedForward network is there to account for assembly of the context vector and the hidden state to produce the output word.
